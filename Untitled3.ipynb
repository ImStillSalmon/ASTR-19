```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
data = np.random.weibull(a, 1000)

```

```python
plt.hist(data, bins=100, color="skyblue", edgecolor="black")

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("weibull random")

plt.savefig("weibull_hist.pdf")
plt.show()
```
